```python
train_df = pd.read_csv("/kaggle/input/train-csv/Train.csv")
```

```python
train_df.isna().sum()
```

```python
X = train_df.drop(['class'], axis=1)
y = train_df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#applying PCA
pca = PCA(n_components=0.95)  
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
#alternatively trying only svd
svd = TruncatedSVD(n_components=26, random_state=42)  # Adjust n_components as needed
X_train_svd = svd.fit_transform(X_train)
X_test_svd = svd.transform(X_test)
```

```python
# Create an SVM classifier
svm_model = SVC(kernel='rbf', random_state=42)  #rbf kernel


svm_model.fit(X_train_svd, y_train)
```

```python
y_pred = svm_model.predict(X_test_svd)


f1 = f1_score(y_test, y_pred, average='weighted')


print(f'F1 Score: {f1}')
```

```python
# Decision Tree classifier
dt_model = DecisionTreeClassifier(random_state=42)


dt_model.fit(X_train_pca, y_train)
```
